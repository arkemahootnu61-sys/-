```python
import numpy as np
feng = np.array([1,2,3,4,5])
print(type(feng))
print(feng)
#就是相当于向量加一
print(feng + 1)
print(feng ** 2)
#构造多维数组
xin = np.array([[1,2,3,4,5]
               ,[2,3,4,5,6]])
print(feng.shape)
print(xin.shape)
```

```python
print(feng.size,xin.size)
```
