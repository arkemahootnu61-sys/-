```python
import numpy as np
feng = np.array([1,2,3,4,5])
print(type(feng))
print(feng)
#就是相当于向量加一
print(feng + 1)
print(feng ** 2)
#构造多维数组
xin = np.array([[1,2,3,4,5]
               ,[2,3,4,5,6]])
print(feng.shape)
print(xin.shape)
```

```python
print(feng[2])
```

```python
#查看向量数字的类型,只能是相同类型，如果不是相同类型，会向下兼容
print(feng.dtype)
print(xin.dtype)
```

```python
print(feng.itemsize)
```

```python
print(feng.size,xin.size)
```

```python
#查看维度
print(feng.ndim,xin.ndim)
```

```python
#填充为0
xin.fill(0)
print(xin)
```

```python
#切片还有索引
print(feng[2:],feng[-2:-1])
print(xin[1:2])
#多维的切片，中间用逗号隔 开
print(xin[0:1,3:])
print(xin[1,2])
#还可以赋值修改
xin[1,2] = 10000
print(xin)
```

```python
zeng = feng
print(zeng)
zeng[1] = 10
print(feng)
print(zeng)
#说明这个zeng和feng都是指向同一个地址
#怎么解决,使用深拷贝
zeng = feng.copy
```

```python
#构造等差数组,左开右闭的
feng = np.arange(1,10,2)
print(feng)
mask = np.array([1,0,0,0,0,0,1],dtype = bool)
print(mask)
```

```python
#随机数,从0到1中间随机出十个数字
random_num = np.random.rand(10)
print(random_num)
```
